```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
days = np.arange(1,101)
initial_price = 100
daily_returns = np.random.normal(loc=0,scale=1,size=100)
stock_prices = initial_price + np.cumsum(daily_returns)
```

```python
def moving_average(prices, window_size):
    return np.convolve(prices, np.ones(window_size)/window_size, mode='valid')

sma_10 = moving_average(stock_prices, 10)
sma_30 = moving_average(stock_prices, 30)
```

```python
volatility = np.std(stock_prices)
print(f"Stock Price Volatility: {volatility:.2f}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(days, stock_prices, label="Stock Price", color='blue', alpha=0.6)
plt.plot(days[9:], sma_10, label="10-Day SMA", color='red', linestyle='dashed')
plt.plot(days[29:], sma_30, label="30-Day SMA", color='green', linestyle='dashed')

plt.xlabel("Days")
plt.ylabel("Stock Price ($)")
plt.title("Stock Price Visualization with Moving Averages")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(["Stock Volatility"], [volatility], color='purple', alpha=0.7)
plt.title("Stock Price Volatility")
plt.ylabel("Standard Deviation of Prices")
plt.show()
```
